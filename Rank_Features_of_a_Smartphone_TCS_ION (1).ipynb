#Model creation

```python
data.dtypes
```

```python
#all the object types has to be dropped.
data=data.drop(['isbluetooth','isdual','is4g','is3g','ist','isw','cores'],axis=1)
```

```python
data.dtypes
```

```python
#spliting dataset
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

```python
# Gausian classifier

from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
```

```python
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
y_pred = dt_model.predict(x_test)
```

```python
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
```

```python
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

```python
#SVM model
#linear

from sklearn.svm import SVC
svm_linear = SVC(kernel = 'linear')
svm_linear.fit(x_train, y_train)
y_pred = svm_linear.predict(x_test)
```

```python
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

```python
#poly kernel
svm_linear = SVC(kernel = 'poly', degree = 3)
svm_linear.fit(x_train, y_train)
y_pred = svm_linear.predict(x_test)
```

```python
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

```python
#rbf kernal
svm_linear = SVC(kernel = 'rbf')
svm_linear.fit(x_train, y_train)
y_pred = svm_linear.predict(x_test)
```

```python
print("accuracy score is {}%".format(accuracy_score(y_pred,y_test)))
print("confusion matrix is {}%".format(confusion_matrix(y_pred,y_test)))
print("{}".format(classification_report(y_pred,y_test)))
```

Logistic Regression has the most accurate Modelling method

#mutual_info_classification




```python
from sklearn.feature_selection import mutual_info_classif
mu=mutual_info_classif(X,Y)
mu=pd.Series(mu,index=X.columns)
mu.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,8))
mu.sort_values(ascending=False).plot.bar()
```

From the bar graph, we can see that ram is the most important feature as it has the highest dependency, so we will consider it while building the model and can also remove those they have no dependency.


#Chi2 Test

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
scores,p_values=chi2(X,Y)

p_value=pd.Series(p_values,index=X.columns)
p_value.sort_values()
```

```python
plt.figure(figsize=(12,8))
p_value.sort_values().plot.bar()
```

THE LESSER THE VALUE OF P THE MORE IMPORTANT IS THE FEATURE

With the help of the above two methods, we can select the important features.
